```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Colored Scatterplot

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.show()
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-virginica':2,'Iris-versicolor':1})
```

```python
# to color the different species 
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.colorbar()
plt.show()
# through cmap= can apply different themes
```

```python
# to color the different species 
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.3)
plt.colorbar()
plt.show()
# through alpha = 'numeric' can handle opacity
```

# Plot size

```python
plt.figure(figsize=(15,7)) #(width,height)
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
plt.show()
```

# Annotation
- with this we can place text on graph

```python
batters = pd.read_csv("batter.csv")
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.xlabel('avg')
plt.ylabel("strike_rate")
plt.show()
```

```python
x = [1,2,3]
y = [2,3,4]
plt.scatter(x,y)
plt.text(1,2,"First")
plt.text(2,3,"Second")
plt.text(3,4,"Third")
```

**Now, apply with for loops**

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
# using s that is size
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# **``Horizontal and Vertical Line``**

# want to see batter who have strike_rate greater than 130

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.axhline(130,color='red')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.axhline(130,color='red')
plt.axvline(30,color='red')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# Benefits of using subplots
- you can create multiple graphs side by side

```python
# now subplot way

fig ,ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('Something')
ax.set_xlabel('average')
ax.set_ylabel('Strike')

#fig.show()
```

```python
plt.subplots(nrows=2,ncols=1)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1)

ax[0].scatter(batters['avg'],batters['strike_rate'])
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title("Avg Vs Strike_Rate")
ax[0].set_ylabel("Strike_rate")

ax[1].set_title("Avg Vs Runs")
ax[1].set_ylabel("Runs")
ax[1].set_xlabel("Avg")
```

**In above my x-axis on both ``axes`` are same that is ``Avg``**
- so, we will use ``sharex=True``

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=((15,9)))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='green')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title("Avg Vs Strike_Rate")
ax[0].set_ylabel("Strike_rate")

ax[1].set_title("Avg Vs Runs")
ax[1].set_ylabel("Runs")
ax[1].set_xlabel("Avg")
```

```python
fig ,ax =  plt.subplots(nrows=2,ncols=2,figsize=((18,9)))

ax[0][0].scatter(batters['avg'],batters['runs'],c='red')
ax[1][0].hist(batters['avg'])
ax[0][1].scatter(batters['avg'],batters['strike_rate'],c='g')
ax[1][1].hist(batters['runs'])
```

**Effective way to create subplots**

```python
fig = plt.figure(figsize=((10,8)))

ax1 = fig.add_subplot(2,1,1) #(no.ofrows , no.ofcols , first graph...)
ax1.scatter(batters['avg'],batters['runs'],color='orange')

ax2 = fig.add_subplot(2,2,3)
ax2.hist(batters['avg'])

ax3 = fig.add_subplot(2,2,2)
ax3.scatter(batters['avg'],batters['strike_rate'],color='green')

ax4 = fig.add_subplot(2,2,4)
ax4.hist(batters['strike_rate'])
```

# 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2  #(x^2 + y^2)
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z,cmap='winter')
fig.colorbar(p)
```

**2nd Graph**

```python
z = np.sin(xx) + np.cos(yy)
```

```python
fig = plt.figure()

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap='summer')
plt.colorbar(p)
```

3rd graph

```python
z = np.sin(xx) + np.log(yy)
```

```python
fig = plt.figure()

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap='winter')
plt.colorbar(p)
```

# Counter Plot
- it converts 3d plot into 2d representation
- it's a 2d plot

```python
z = xx**2 + yy**2  #(x^2 + y^2)

fig = plt.figure()

ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='winter')
```

```python
z = xx**2 + yy**2  #(x^2 + y^2)

fig = plt.figure()

ax = plt.subplot()
p = ax.contourf(xx,yy,z)
```

```python

z = np.sin(xx) + np.cos(yy)
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
#2nd 2d version

z = np.sin(xx) + np.cos(yy)
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
plt.colorbar(p);
#blue point is 'gadhaa' and yellow points are high
```

# Heatmap
- usually creates with seaborn 
- but for a time let's do with
- matplotlib using pivot_table()

```python
delivery=pd.read_csv("IPL_Ball_by_Ball_2008_2022.csv")
delivery.head()
```

```python
temp = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
# har over ki har bol par sixes
grid=temp.pivot_table(index='overs',columns=['ballnumber'],values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=((10,7)))
plt.imshow(grid)

plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))

plt.colorbar()
```
