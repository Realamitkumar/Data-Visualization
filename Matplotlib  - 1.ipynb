```python
# plt.scatter() 
batter = pd.read_csv("batter.csv")
batter = batter.head(50)
# avg --> reliable batter
# strike_rate --> most destructive batter
plt.scatter(batter['avg'],batter['strike_rate'],marker='+')
plt.title("Avg and SR analysis of Top 50 Batsman")
plt.xlabel("Avg")
plt.ylabel("SR")
```

```python
# size
import seaborn as sns
tips = sns.load_dataset("tips")
tips
plt.scatter(tips["total_bill"],tips['tip'],s=tips['size']*20)
plt.title("total_bill and tip analysis , along with size")
plt.xlabel("total_bill")
plt.ylabel("tip")
```

```python
# scatter ... using plt.plot()
# faster technique

plt.plot(tips["total_bill"],tips['tip'],'o')
```

### Bar chart

image.png







- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black');
```

```python
plt.barh(colors,children,color='black');
```

```python
# color and label 
df = pd.read_csv('batsman_season_record.csv')
plt.bar(df['batsman'],df['2015'],width=0.5)
plt.bar(np.array(df.shape[0]),df['2015'])
```

**To see different year chart ``side by side``**

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width = 0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width = 0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
np.arange(df.shape[0])
```

```python
# multiple bar chart
children = [10,20,40,10,30]
colors = ['red red red ','blue blue blue ','green green green','yellow yellow yellow','pink pink pink']

plt.bar(colors,children,color='black')
plt.xticks(rotation = 20)
plt.show()
```

**important**

```python
# stacked bar chart
plt.bar(df['batsman'],df['2017'],color='r',label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],color='y',label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']),color='g',label='2015')
plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
df = pd.read_csv('gayle-175.csv')
plt.pie(df['batsman_runs'],labels=df.batsman,autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show();
```

```python
#explode
plt.pie(df['batsman_runs'],labels=df.batsman,autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1])
plt.show()
```

```python
#shadow
plt.pie(df['batsman_runs'],labels=df.batsman,autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
