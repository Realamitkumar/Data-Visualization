Each plotting function in seaborn is either an Axes-level function or a figure-level function. An Axes-level function draws onto a single matplotlib Axes and does not effect the rest of the figure. A figure-level function, on the other hand, controls the entire figure.Hers is a quick summary for Axes-level and Figure-level functions

### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',style='time',size='size')


#OR
#sns.scatterplot(tips['total_bill'],tips['tip'],style=tips['time'],hue=tips['sex'],size=tips['size'])
```

```python
# scatter plot -> figure level function--> rectangular
# seaborn suggest to create scatter plot using figure level 
sns.relplot(data=tips,x='total_bill', y='tip',hue='sex',style='smoker',size='size')
```

```python
# also create different plot using ``relplot(kind='')``
# default='scatterplot'
sns.relplot(data=tips,x='total_bill', y='tip',kind='line')
```

```python
# line plot
gap = px.data.gapminder()
temp=gap[gap['country']=='India']
```

```python
sns.lineplot(data=temp,x='year',y='lifeExp')
```

```python
# using figure level func.. that is relplot
sns.relplot(data=temp,x='year',y='lifeExp',kind='line')
```

```python
# hue -- style
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country')
```

**little bit different from above graph**

- legend was outside when you used relplot() to create line plot

- place legend outside and becoz he was the owner of the containor

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
# it quiete obivous becoz it's a figure level 
# so have to use figure level function i.e., relplot()
sns.relplot(data=tips,x='total_bill',y='tip',col='sex')
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
#same thing with figure level func.. that is displot()
sns.displot(data=tips,kind='hist',x='total_bill')
```

```python
# bins parameter
sns.displot(data=tips,x='total_bill',kind='hist',bins=10)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# kaide se this is your ''countplot''
sns.displot(data=tips, x='day', kind='hist') 
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex') 
```

```python
# element parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex') 
```

```python
#2nd dataset
titanic = sns.load_dataset('titanic')
```

```python
sns.displot(data=titanic, x='age', kind='hist',hue='sex',element='step') 
```

```python
# faceting using col and row -> not work on histplot function becoz that is
# axes level funn and this work with figure level function
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
```

```python

# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
# will discuss in stats
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data = tips, x = 'total_bill', y = 'tip',kind = 'hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data = tips, x='total_bill', y='tip',fill=True)
# dark whole is more dense area 
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
g=gap.pivot(index='country',columns='year',values='lifeExp')

#or
#g = gap.pivot("country", "year", "lifeExp")
#sns.heatmap(g)


# axes level function
sns.heatmap(g)
```

```python
#annot
g = gap[gap['continent']=='Europe'].pivot('country','year','lifeExp')

plt.figure(figsize=((17,15)))
sns.heatmap(g,annot=True,cmap='viridis',linewidth=0.5)
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
